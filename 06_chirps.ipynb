```python
# CHIRPS monthly global
downloaded = True

extent = '-46.200579418 -11.200452249 -44.927499105 -12.210613802' # around the Bahia farm
#extent = '-82.138213125 12.460529649 -34.636212417 -49.783471279' # South America
#extent = '-81.533404133 6.658986797 -33.888358209 -34.958986797' # Brazil

# mode: monthly dekad pentad daily
mode = 'monthly'
where = 'south_america'

if downloaded == False:
    base = 'https://data.chc.ucsb.edu/products/CHIRPS-2.0/global_monthly_EWX/data/'
else:
    base = '/home/geodata/Raster/chirps/'
    
out = f'/home/geodata/Raster/chirps/{mode}_{where}/'

Path(f'{out}').mkdir( parents = True, exist_ok = True)
out
```

```python
files = sorted(glob(base+'*.tiff'))
```

```python
anos = np.unique([x.split('.')[-3] for x in files])
meses = np.unique([x.split('.')[-2] for x in files])
```

```python
anos
```

```python
# the downloaded way
files = sorted(glob(base+'*.tiff'))
for a in anos:
    for m in meses:
        f = f'{base}/data.{a}.{m}.tiff'
        try:
            os.system(f'gdal_translate -projwin {extent} -of GTiff {f} {out}{a}-{m}.tif')
        except:
            print(f'{a}-{m} not in the dataset. ')
```

```python
# the download way
os.chdir(out)
for a in np.arange(1985,2025):
    for m in [str(x).zfill(2) for x in np.arange(1,13)]:
        print(a,m)
        try:
            os.system(f'wget {base}data.{a}.{m}.tiff --no-verbose')
            print(f'downloaded')
            os.system(f'gdal_translate -projwin {extent} -of GTiff data.{a}.{m}.tiff {a}.{m}.tif')
            os.system(f'mv data.{a}.{m}.tiff ..')
        except:
            print(f'{a}-{m} not in the dataset. ')
```
